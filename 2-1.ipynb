```python
import numpy as np

def euler(df, y0, x0, dx, n) -> np.array:
    x = np.array([x0])
    y = np.empty((0, len(y0)))
    y = np.vstack((y, np.array(y0)))
    for _ in range(n):
        xi = x[-1] + dx
        yi = y[-1, :] + df(x[-1], y[-1,:])*dx
        x = np.vstack((x, np.array([xi])))
        y = np.vstack((y, np.array(yi)))
    return np.hstack((x, y))
```

```python
x0 = 0
y0 = np.array([20, 10])
df = lambda x, y: np.array([-y[1], -y[0]])
n = 6
dx = 0.1

xy = euler(df, y0, x0, dx, n)
print(xy.shape)
```

```python
import matplotlib.pyplot as plt

plt.plot(xy[:, 0], xy[:, 1])
plt.plot(xy[:, 0], xy[:, 2])
```
